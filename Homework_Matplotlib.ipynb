```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "mouse_drug_data.csv"
clinical_trial_data_to_load = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data=pd.read_csv(mouse_drug_data_to_load)
clinic_data=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
data = pd.merge(clinic_data, mouse_data, how="left", on="Mouse ID")
data =pd.DataFrame(data)

```

## Tumor Response to Treatment

```python
tumor_data = pd.DataFrame(data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean())
tumor_data = tumor_data.reset_index()
tumor_data=tumor_data[~tumor_data['Drug'].str.contains('Ceftamin')]
tumor_data=tumor_data[~tumor_data['Drug'].str.contains('Naftisol')]
tumor_data=tumor_data[~tumor_data['Drug'].str.contains('Propriva')]
tumor_data=tumor_data[~tumor_data['Drug'].str.contains('Ramicane')]
tumor_data=tumor_data[~tumor_data['Drug'].str.contains('Stelasyn')]
tumor_data=tumor_data[~tumor_data['Drug'].str.contains('Zoniferol')]

tumor_data.head(12)
tumor_data1 = pd.DataFrame(data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem())
tumor_data1 = tumor_data1.reset_index()
fig, ax = plt.subplots()
for label, d in tumor_data.groupby('Drug'):
    color = list(np.random.uniform(size=3))
    d.plot.scatter("Timepoint", "Tumor Volume (mm3)", yerr=tumor_data1,
label=f" {label}", title=("Tumor Volume Response to different treatments over time"), ax=ax, c=[color], figsize=(14,14), s=80)
```
